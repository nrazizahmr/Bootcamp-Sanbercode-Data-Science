# Data Structure

<h2>Soal 1: Perbedaan Data Structure</h2>

Jawab Pertanyaan di bawah ini:

Jelaskan perbedaan dari List, Tuple, Set dan Dictionary!

List, Tuple, Set, dan Dictionary adalah empat jenis struktur data di Python yang memiliki karakteristik berbeda sesuai dengan kebutuhan penggunaannya.


*   List adalah struktur data yang paling fleksibel,memungkinkan kita mengakses elemen menggunakan indeks secara berurutan (sequential) dan memanipulasi elemen karena sifatnya yang mutable. List juga mendukung elemen yang duplikat dan bersifat iterable sehingga cocok untuk menyimpan data dalam urutan tertentu.
*   Tuple, di sisi lain, memiliki sifat yang lebih kaku karena bersifat immutable, artinya elemen di dalamnya tidak dapat diubah setelah tuple dibuat. Namun, seperti List, Tuple juga mendukung pengaksesan elemen secara berurutan dan dapat menyimpan elemen yang duplikat. Tuple sering digunakan ketika data tidak boleh diubah.
*   Berbeda dengan List dan Tuple, Set adalah struktur data yang hanya menyimpan elemen-elemen unik, sehingga tidak mengizinkan adanya elemen yang duplikat. Set tidak mendukung pengaksesan elemen berdasarkan indeks karena bersifat non-sequential, tetapi bersifat mutable, memungkinkan elemen untuk ditambah atau dihapus. Set memiliki karakteristik seperti himpunan dalam matematika, misalnya mendukung operasi seperti union, intersection, dan difference, tetapi elemen-elemen dalam Set tidak dijamin berurutan.
*   Sementara itu, Dictionary adalah struktur data yang menyimpan pasangan key-value. Tidak seperti List, Tuple, dan Set yang hanya menyimpan elemen tunggal, Dictionary mengasosiasikan setiap elemen (value) dengan sebuah kunci unik (key). Key dalam Dictionary harus bersifat immutable (seperti string, angka, atau tuple), tetapi value bisa berupa tipe data apa pun, termasuk struktur data lainnya. Dictionary bersifat mutable, sehingga pasangan key-value dapat ditambah, dihapus, atau diubah. Elemen dalam Dictionary tidak berurutan (non-sequential), tetapi dapat diakses secara langsung menggunakan key, bukan indeks. Dictionary sangat berguna untuk menyimpan data yang memiliki hubungan atau asosiasi, seperti data berbentuk tabel.


Perbedaan utama di antara keempatnya terletak pada struktur penyimpanan, cara pengaksesan elemen (indeks vs key), sifat mutable/immutable, dan dukungan elemen duplikat. List dan Tuple cocok untuk data yang berurutan, Set untuk data unik tanpa urutan, sedangkan Dictionary ideal untuk data dengan hubungan key-value.

## Soal 6: List Comprehension

Gunakan metode **list comprehension** untuk mencari anggota dari S yang habis di bagi 2, kemudian assign hasilnya dalam bentuk list ke dalam variabel T.

```python
S = [0, 1, 4, 9, 16, 25, 36, 49, 64, 81]

T =[i for i in S if i % 2==0]

print(T)
```

Expected Output:

[0, 4, 16, 36, 64]

<h2>Soal 10: Nested Dictionary</h2>
    
Lengkapi kode untuk menghasilkan suatu output yang di harapkan

```python
country = {
           'spain': { 'capital':'madrid', 'population':46.77 },
           'france': { 'capital':'paris', 'population':66.03 },
           'germany': { 'capital':'berlin', 'population':80.62 },
           'norway': { 'capital':'oslo', 'population':5.084 }
         }

# berapa populasi dari kota german?
populasi=country['germany']['population']
print(populasi)
# update data baru, yaitu negara indonesia dengan capital jakarta dan poulasi 250
country['indonesia']={ 'capital':'jakarta', 'population':250 }

print(country)
```

<h2>Soal 11: Loop Dictionary</h2>
    
Lengkapi kode untuk menghasilkan suatu output yang di harapkan

```python
country = {
           'spain': { 'capital':'madrid', 'population':46.77 },
           'france': { 'capital':'paris', 'population':66.03 },
           'germany': { 'capital':'berlin', 'population':80.62 },
           'norway': { 'capital':'oslo', 'population':5.084 },
           'indonesia' : {'capital':'jakarta', 'population':250}
         }

for country_name, details in country.items():
    print('Ibukota '+country_name+' adalah ' + details['capital'])
```

Expected Output:
    
Ibukota spain adalah madrid

Ibukota france adalah paris

Ibukota germany adalah berlin

Ibukota norway adalah oslo

Ibukota indonesia adalah jakarta

## Soal 12: Remove Duplicate using set


Hilangkan nilai duplikat dari sebuah objek list dengan menggunakan cara set sehingga menjadi sebuah tipe set

```python
obj_list = [1, 2, 3, 4, 5, 5, 6, 7, 8, 9, 9, 10]

#using set
duplikat = set(obj_list)

print(duplikat)
```

## Soal 13: Mengubah dan menghapus anggota set

Ubahlah dan hapus anggota set sehingga mendapatkan hasil yang diinginkan


```python
set_1 = {1, 2, 3, 4, 5}

#tambahkan anggota set dengan nilai {6,7,8}
set_5 = {6,7,8}
union=set_1 | set_5

#hapus nilai anggota set 8
union.remove(8)
print(union)
```

Expected output:

{1, 2, 3, 4, 5, 6, 7}

## Soal 14: Operasi pada Set

Carilah irisan dari ke dua set dengan menggunakan metode intersection

```python
set_2 = {6, 8, 9, 10, 24}
set_3 = {6, 10, 8, 25, 13}

#intersection set
inter_result=set_2 & set_3
sorted_inter=sorted(inter_result)
print(sorted_inter)
```

Expected output:


{6, 8, 10}
