```python
x = 10
y = 20

#and
if x > 5 and y < 30:
    print("Kedua kondisi benar!")
else:
    print("Salah satu atau kedua kondisi salah")

#or
if x > 5 or y < 10:
    print("Satu kondisi benar")
else:
    print("Kedua kondisi salah")

#not
if not (x < 5):
    print("Kondisi x < 5 tidak benar")
else:
    print("Kondisi x < 5 benar")
```

<h2>Soal 4: Comparison Operator dengan fungsi</h2>

Buatlah sebuah fungsi yang menerima satu argument bertipe data numeric dan menghasilkan sebuah return sebagai berikut :
- menghasilkan return 'Aneh' jika nilai dari argument tersebut adalah bilangan ganjil
- menghasilkan return 'Tidak Aneh' jika nilai dari argument tersebut adalah bilangan genap dan diantara nilai 2 sampai 5 (2 dan 5 termasuk)
- menghasilkan return 'Aneh' jika nilai dari argument tersebut adalah bilangan genap dan diantara nilai 6 sampai 20 (6 dan 20 termasuk)
- menghasilkan return 'Tidak Aneh' jika nilai dari argument tersebut adalah bilangan genap dan lebih besari dari 20

```python
def cek_angka(n):
    if n%2 != 0:
        return 'Aneh'
    elif n%2 == 0 and 2<=n<=5:
        return 'Tidak Aneh'
    elif n%2 == 0 and 6<=n<=20:
        return 'Aneh'
    elif n%2 == 0 and n>20:
        return 'Tidak Aneh'

print(cek_angka(3))
print(cek_angka(4))
print(cek_angka(10))
print(cek_angka(22))
```

<h2>Soal 5: While Loop dan For Loop</h2>

- Apa perbedaan while loop dan for loop?
- Berikan contoh sederhana cara menggunakan while loop dan for loop

Perbedaan

*   Cara Kerja: While loop berjalan selama kondisi tertentu bernilai True. For loop berjalan untuk setiap elemen dalam objek yang dapat diiterasi (iterable).
*   Penggunaan Utama: While loop digunakan jika jumlah iterasi tidak diketahui sebelumnya, bergantung pada kondisi dinamis. For loop digunakan jika jumlah iterasi diketahui sebelumnya atau saat bekerja dengan iterable.
*   Kondisi Berhenti: While loop memerlukan kondisi eksplisit untuk berhenti tanpa kondisi yang tepat, dapat menyebabkan infinite loop. For loop berhenti secara otomatis setelah semua elemen dalam iterable selesai diiterasi.
*   Kesederhanaan: While loop lebih kompleks karena memerlukan pengelolaan variabel kontrol dan kondisi manual. For loop lebih sederhana karena iterasi dikelola secara otomatis.
*   Jenis Data: While loop tidak bergantung pada iterable, hanya bergantung pada kondisi. For loop membutuhkan objek iterable seperti list, tuple, range, atau string.
*   Kapan Digunakan: While loop digunakan untuk pengulangan yang tidak ditentukan jumlah iterasinya sebelumnya. For loop digunakan untuk iterasi yang terdefinisi dengan jelas atau bekerja dengan elemen dalam iterable.







```python
# jawaban contoh
# while loop
count = 1
while count <= 5:
    print(f"While loop: {count}")
    count += 1

# for loop
for count in range(1, 6):
    print(f"For loop: {count}")
```

<h2>Soal 6: While Loop</h2>

- Dalam setiap iterasi update nilai variable inputan itu, jika input lebih dari 0, maka kurangi variabel tersebut sebesar 1, selain itu tambahkan 1
- print nilai variable input dalam setiap awal iterasi

```python
def fungsi_while(angka):
  if angka>0:
      while angka >=1:
        print(angka)
        angka -= 1
  elif angka<0:
      while angka < 0:
        print(angka)
        angka += 1
a = 10
print("Jika a = 10")
fungsi_while(a)
print()
a=-10
print("Jika a = -10")
fungsi_while(a)
```

<h2>Soal 7: For Loop</h2>

Lengkapi kode untuk menghasilkan suatu output yang diharapkan:





*   Cari siswa mana saja yang memiliki nilai lebih dari sama dengan 80. Masukkan kedalam sebuah list. print hasilnya






```python
data_siswa = [
    {
        "nama":"Budi",
        "nilai": 90
    },
    {
        "nama":"Nina",
        "nilai": 78
    },
    {
        "nama":"Rudi",
        "nilai": 91
    },
    {
        "nama":"Olivia",
        "nilai": 76
    },
    {
        "nama":"Leo",
        "nilai": 80
    },
    {
        "nama":"Liam",
        "nilai": 67
    },
    {
        "nama":"Sheila",
        "nilai": 76
    }
]
```

```python
#code here
siswa_up_80=[]
for siswa in data_siswa:
    if siswa["nilai"] >= 80:
      siswa_up_80.append(siswa["nama"])

print(siswa_up_80)
```

['Budi', 'Rudi', 'Leo']

- Buatlah sebuah loop dengan mengiterasi sebuah objek list kemudian di kalikan dengan index dari list tersebut dan print hasilnya
