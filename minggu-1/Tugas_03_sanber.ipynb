<h2>Soal 2: Built-in Function</h2>

Jawab Pertanyaan di bawah ini:
    
Apa itu built-in Function?

Sebutkan 3 built-in Function di pyhon!

Berikan contoh masing2 1 cara penggunaan built-in function yang kamu sebutkan sebelumnya

//isi jawaban text disini


*   len(): Digunakan untuk menghitung jumlah elemen dalam objek yang dapat dihitung, seperti list, string, tuple, atau dictionary.
*   type(): Digunakan untuk mengetahui tipe data dari suatu objek.
*   max(): Digunakan untuk menemukan nilai terbesar dalam sekumpulan data seperti list, tuple, atau string.

```python
# Contoh penggunaan built in function
mylist=[2,9,7,4,0]
print('List : ' + str(len(mylist)))

myvar=22.7
print('Type : ' + str(type(myvar)))

mynumber=[9, 20, 74, 1, 5, 43]
print('Max : ' + str(max(mynumber)))
```

<h2>Soal 3: Method dan Function</h2>

Jawab Pertanyaan di bawah ini:
    
- Apa perbedaan method dan function?



*   Function: Merupakan blok kode yang berdiri sendiri dan dirancang untuk menjalankan tugas tertentu. Function dapat dipanggil di mana saja dalam kode, asalkan didefinisikan sebelumnya atau merupakan fungsi bawaan Python.

Keduanya digunakan untuk mempermudah pengolahan data di Python, tetapi method lebih spesifik karena bergantung pada objek tertentu.





<h2>Soal 4: Menggunkan Method String</h2>

Lengkapi kode untuk menghasilkan suatu output yang di harapkan

```python
kalimat = "ayo belajar data science"

# gunakan method untuk mengubah nilai kalimat menjadi uppercase semua kemudian tampilkan hasilnya
capital=kalimat.upper()
print(capital)

# gunakan method untuk menghitung berapa huruf a di dalam kalimat
hitung_huruf=kalimat.count('a')
print(hitung_huruf)
```

Buatlah suatu fungsi yang menerima satu input argumen berbentuk list dan mempunyai elemen bertipe numeric semua, dimana fungsi tersebut berguna untuk menghitung rata2 dari kumpulan elemen list tersebut. namai fungsi tersebut 'mean_list'

```python
obj_list = [11.25, 18.0, 20.0, 10.75, 9.50, 13.45, 23.0, 7.0, 8.45, 2.50]
def mean_list(inp_list):
 n=len(obj_list)
 inp_list=sum(obj_list)
 mean=inp_list/n
 return mean

print(mean_list(obj_list))
```

# Soal 7. Mapping Fungsi lambda

Hitunglah jumlah bilangan genap dari sebuah objek list berikut menggunakan lambda function

```python
obj_list = [i for i in range(1000)]
genap = list(filter(lambda x: x % 2 == 0, obj_list))
hasil = len(genap)

print(hasil)
```

Expected output:

500
