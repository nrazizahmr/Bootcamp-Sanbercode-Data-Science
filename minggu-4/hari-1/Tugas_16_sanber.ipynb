# Soal 1. Menentukan jenis data categorical atau numeric
Jelaskan data kategorikal (kualitatif) dan data numerik (kuantitatif)



Jawab disini:

Data kategorikal adalah data yang mengelompokkan objek atau individu ke dalam kategori tertentu tanpa nilai numerik yang bermakna dalam perhitungan. Misalnya, ketika kita berbicara tentang warna mobil—merah, biru, atau hitam—kita sedang menggunakan data kategorikal. Data ini juga dapat diklasifikasikan lebih lanjut menjadi data nominal dan data ordinal. Data nominal tidak memiliki urutan tertentu, seperti jenis kelamin (pria atau wanita) atau jenis pekerjaan (dokter, insinyur, guru). Sementara itu, data ordinal memiliki tingkatan atau hierarki, seperti tingkat kepuasan pelanggan (sangat puas, puas, netral, tidak puas) atau tingkat pendidikan (SD, SMP, SMA, S1).

---

Data numerik adalah data yang berupa angka dan dapat dihitung serta dianalisis menggunakan operasi matematika. Data ini terbagi menjadi dua jenis utama, yaitu data diskrit dan data kontinu. Data diskrit adalah data yang hanya dapat memiliki nilai tertentu dan tidak bisa dibagi lebih kecil, seperti jumlah anak dalam keluarga (1, 2, 3) atau jumlah mobil di tempat parkir (10, 15, 20). Sementara itu, data kontinu adalah data yang bisa memiliki nilai pecahan dalam suatu rentang tertentu, seperti berat badan seseorang (65,4 kg), tinggi badan (170,5 cm), atau suhu udara (27,8°C).

```python
# code here
import seaborn as sns

df = pd.read_csv('austin_weather.csv', na_values=['na', '-'])

df['DewPointAvgF'] = df['DewPointAvgF'].astype(float)
df['HumidityAvgPercent'] = df['HumidityAvgPercent'].astype(float)
df['TempAvgF'] = df['TempAvgF'].astype(float)
df['WindAvgMPH'] = df['WindAvgMPH'].astype(float)

sample_df = df.sample(1000, random_state=42)

# Tanpa Handling Missing Value
plt.figure(figsize=(10,6))
sns.scatterplot(
    x=sample_df['HumidityAvgPercent'],
    y=sample_df['DewPointAvgF'],
    hue=sample_df['TempAvgF'],
    size=sample_df['WindAvgMPH'] * 20,
    palette='coolwarm',
    alpha=0.7
)
plt.xlabel('HumidityAvgPercent')
plt.ylabel('DewPointAvgF')
plt.title('Scatter Plot Sebelum Handling Missing Values')
plt.legend()
plt.show()
```

```python
#Menggunakan Handling Missing Value
df.fillna(method='ffill', inplace=True)

sample_df_ffill = df.sample(1000, random_state=42)

plt.figure(figsize=(10,6))
sns.scatterplot(
    x=sample_df_ffill['HumidityAvgPercent'],
    y=sample_df_ffill['DewPointAvgF'],
    hue=sample_df_ffill['TempAvgF'],
    size=sample_df_ffill['WindAvgMPH'] * 20,
    palette='coolwarm',
    alpha=0.7
)
plt.xlabel('HumidityAvgPercent')
plt.ylabel('DewPointAvgF')
plt.title('Scatter Plot Setelah Handling Missing Values')
plt.legend()
plt.show()
```

```python
#Code here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df2['Dist_Taxi'] = df2['Dist_Taxi'].astype(float)
df2['Dist_Market'] = df2['Dist_Market'].astype(float)
df2['Dist_Hospital'] = df2['Dist_Hospital'].astype(float)
df2['Carpet'] = df2['Carpet'].astype(float)
df2['Builtup'] = df2['Builtup'].astype(float)
df2['Rainfall'] = df2['Rainfall'].astype(float)
df2['House_Price'] = df2['House_Price'].astype(float)

df2['Parking'] = df2['Parking'].astype('category')
df2['City_Category'] = df2['City_Category'].astype('category')

# sebelum handling outliers
plt.figure(figsize=(10,6))
sns.boxplot(x=df2['House_Price'])
plt.title('Boxplot Sebelum Handling Outliers')
plt.show()

def handle_outliers(df, column):
    Q1 = df2[column].quantile(0.25)
    Q3 = df2[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    print(f"Batas bawah & batas atas:\n{lower_bound} {upper_bound}")
    df_clean = df2[(df[column] >= lower_bound) & (df2[column] <= upper_bound)].copy()
    return df_clean

# setelah handling outliers
df2 = handle_outliers(df2, 'House_Price')
plt.figure(figsize=(10,6))
sns.boxplot(x=df2['House_Price'])
plt.title('Boxplot Setelah Handling Outliers')
plt.show()
```
