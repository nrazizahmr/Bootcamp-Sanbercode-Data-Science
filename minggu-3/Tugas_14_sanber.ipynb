```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('vgsales.csv')
df.head()
```

<h2>Soal 1: Preparing Data</h2>

Manipulasi data tersebut dengan menggroupkan berdasarkan Genre, kemudian ambil rata2 penjualan untuk Global_Sales berdasarkan kategori Genre.

```python
#code here
mean_genre = df.groupby("Genre")["Global_Sales"].mean()
mean_genre
```

<h2>Soal 2: Perbandingan Kuantitatif Pieplot - Part a whole using Pie plot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menghasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
#code here
plt.figure(figsize=(4, 4))
mean_genre.plot(
    kind="pie",
    autopct="%1.1f%%",
    startangle=90,
    fontsize=8,
)

plt.title("Mean Global Sales by Genre", fontsize=10)
plt.ylabel("")
plt.tight_layout()
plt.show()
```

<h2>Soal 3: Preparing Data</h2>

Manipulasi data tersebut dengan menggroupkan berdasarkan Genre, kemudian ambil rata2 penjualan untuk setiap Region Kecuali Global_Sales berdasarkan kategori Genre.

```python
#code here
mean_region = df.groupby("Genre")[["NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales"]].mean()
mean_region
```

<h2>Soal 4: Perbandingan Kuantitatif Barplot - Grouping Visualisasi Dengan Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menghasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
#code here
import numpy as np

regions = mean_region.columns
genres = mean_region.index
x = np.arange(len(regions))
width = 0.07

fig, ax = plt.subplots(figsize=(12, 6))

for i, genre in enumerate(genres):
    ax.bar(x + i * width, mean_region.loc[genre], width, label=genre)

ax.set_xlabel("Region Sales", fontsize=10)
ax.set_ylabel("Mean Sales", fontsize=10)
ax.set_title("Mean Sales Video Games by Genre", fontsize=10)
ax.set_xticks(x + width * (len(genres) - 1) / 2)
ax.set_xticklabels(regions)
ax.legend(title="Genre", fontsize=10, bbox_to_anchor=(1, 1), loc="upper left")
ax.grid(axis="y", linestyle="--", alpha=0.7)

plt.tight_layout()
plt.show()
```

Amerika Utara (NA): Di wilayah ini, game dengan genre Platform dan Racing cukup populer selain Aksi dan Petualangan.

Eropa (EU): Mirip dengan NA, EU juga menyukai game dengan genre Platform dan Racing, tetapi dengan porsi yang sedikit berbeda.

Jepang (JP): Jepang memiliki preferensi yang unik dengan genre game Role-Playing yang sangat kuat. Genre ini memiliki rata-rata penjualan yang jauh lebih tinggi dibandingkan wilayah lain.

Wilayah Lain: Wilayah lain menunjukkan pola yang lebih beragam, namun secara umum mengikuti tren global dengan dominasi Aksi dan Petualangan.
