<h2>Soal 1: Pemahaman Component Matplotlib</h2>

- Jelaskan apa itu Figure
- Jelaskan apa itu Axis

*   Figure: Sebuah Figure adalah wadah utama atau tempat semua elemen visual diletakkan. Ini mencakup sumbu (axis), label, judul, legenda, dan elemen grafis lainnya.
*   Sebuah Axis adalah area dalam Figure tempat data diplot. Axes mencakup sumbu X, sumbu Y, (dan dalam kasus 3D, sumbu Z), skala, grid, dan elemen-elemen yang berhubungan dengan data itu sendiri.



<h2>Soal 2: Membuat Component Figure dan Axis</h2>

- Buatlah Figure dan Axis kosong tanpa Data
- Plot data di bawah ini dengan 2 cara, yaitu dengan membuat figure dan axis secara explicit dan Implicit

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 20, 100)
y = np.cos(x/2)
```

```python
# membuat Figure dan Axis dengan Data Kosong
fig, ax = plt.subplots()
plt.show()
```

```python
#eksplisit
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x,y)
```

```python
#implicit
fig = plt.figure()
plt.plot(x,y)
```

<h2>Soal 3: Memasukan multiple data ke dalam satu axis</h2>

- Plot ketiga data berikut ke dalam satu axis, dan gunakan seaborn style
- save data tersebut dengan nama file bebas (apa saja) dan tampilkan kembali gambar tersebut.

```python
x = np.linspace(0, 20, 100)
y = np.cos(x/2)
y2 = np.sin(x)
y3 = np.sin(2*x)
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x,y)
ax.plot(x,y2)
ax.plot(x,y3)
plt.show()
```
