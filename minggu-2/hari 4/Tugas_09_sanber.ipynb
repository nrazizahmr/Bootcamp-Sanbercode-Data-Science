<h2>Soal 1: Penggabungan Data</h2>

- Sebutkan perbedaan antara full outer join, inner join, dan left join



*   Full outer join: Full Outer Join menghasilkan gabungan index bari dari setiap data yang di gabung. Full Outer Join menggabungan pula column yang berbeda serta memberikan nilai null untuk nilai column yang tidak di miliki oleh suatu index tertentu.
*   Inner join: Inner join ini akan menggabungkan data antara nilai key/referensi yang beririsan.
*   Left Join: Left join menghasilkan seluruh data dari data yang kiri ditambah data kanan yang memiliki nilai key yang sama dengan data dari kiri.

<h2>Soal 3: Penggabungan Data</h2>

Lakukan jenis penggabungan data full outer join, inner join, left outer join dan right outer join pada kedua data berikut:

```python
import pandas as pd

data1 = {
    'id' : [1, 2, 3, 4],
    'Feature1' : ['a1', 'a2', 'a3', 'a4'],
    'Feature2' : ['b1', 'b2', 'b3', 'b4']
}

data2 = {
    'id' : [3, 4, 5 ,6, 7],
    'Feature3' : ['c1', 'c2', 'c3', 'c4', 'c5']
}

df1= pd.DataFrame(data1)
df2= pd.DataFrame(data2)

print(df1)
print(df2)
```

```python
#join data
df_outer_join = pd.merge(df1,df2, on='id', how='outer')
print(df_outer_join)

df_inner_join = pd.merge(df1,df2, on='id', how='inner')
print(df_inner_join)

df_left_join = pd.merge(df1,df2, on='id', how='left')
print(df_left_join)

df_right_join = pd.merge(df1,df2, on='id', how='right')
print(df_right_join)
```

# Data GroupBy

<b>Note</b>: Pada kuis kali ini kita akan menggunakan data diamonds.csv yang bisa kalian download [disini](https://drive.google.com/uc?export=download&id=1qQiM9utQUThTPq8jk8stPMh_y2zfqYOd). setelah di download, kalian pindahkan data tersebut satu level dengan direktori file notebook ini.

```python
import pandas as pd
df = pd.read_csv('diamonds.csv')
df.drop('Unnamed: 0', axis=1, inplace=True)
```

<h2>Soal 5: Process Grouping</h2>

Jelaskan prosess yang terjadi dalam grouping data menggungkan method <b>groupby()</b> dari pandas.

jawab:

Proses grouping data menggunakan method groupby() dari Pandas melibatkan beberapa tahapan logis yang memungkinkan data dikelompokkan berdasarkan nilai tertentu, kemudian operasi seperti agregasi, transformasi, atau filtrasi dapat dilakukan pada setiap kelompok secara terpisah. Berikut penjelasan rinci prosesnya,


*   Agregasi: Operasi yang mengembalikan satu nilai per kelompok (misalnya, sum, mean, count).
*   Transformasi: Operasi yang mengembalikan array dengan ukuran yang sama seperti data asli (misalnya, apply, transform).
*   Filtrasi: Operasi untuk menyaring kelompok tertentu (misalnya, filter).


<h2>Soal 6: Grouping All Column</h2>
    
Group data diamond berdasarkan cut (kualitas), lakukan aggregasi dengan fungsi rata2.


```python
#code herereu
result = df.groupby('cut').agg({
    'carat': 'mean',
    'depth': 'mean',
    'table': 'mean',
    'price': 'mean',
    'x': 'mean',
    'y': 'mean',
    'z': 'mean'
})

print(result)
```

<h2>Soal 7: Grouping Some Column base on Multiple Criteria</h2>
    
Lakukan groupby pada data berdasarkan kriteria cut (kualitas) dan color (warna) secara berurutan, dengan aggregasi pada column carat, table, dan price saja. dengan fungsi agregasi rata-rata.

```python
#code here
result = df.groupby(['cut','color']).agg({
    'carat': 'mean',
    'table': 'mean',
    'price': 'mean'
})
result
```

<h2>Soal 8: Group By Multiple Aggregate Base on Multiple Criteria</h2>
    
Lakukan groupby pada data berdasarkan kriteria cut(kualitas) dan color (warna) secara berurutan kemudian aggregasi objek group by nya dengan fungsi rata2 dan median.

```python
#code here
result = df.groupby(['cut', 'color']).agg({
    'carat': ['mean', 'median'],
    'depth': ['mean', 'median'],
    'table': ['mean', 'median'],
    'price': ['mean', 'median'],
    'x': ['mean', 'median'],
    'y': ['mean', 'median'],
    'z': ['mean', 'median']
})
result
```

<h2>Soal 9: Group By With Different Aggregation for specific Column</h2>
    
Lakukan groupby pada data berdasarkan criteria cut(kualitas), kemudian aggreagate dengan fungsi max pada column price dan fungsi min pada column carat.

```python
#code here
result = df.groupby('cut').agg({
    'price': 'max',
    'carat': 'min'
})

print(result)
```

<h2>Soal 10: Group By With Custom Function</h2>
    
Lakukan groupby pada data berdasarkan criteria cut(kualitas), kemudian lakukan aggregasi dengan fungsi standar deviasi pada column carat dan fungsi rentang nilai antara max dan min pada column price.

```python
#code here
result = df.groupby('cut').agg({
    'price': lambda x: x.max() - x.min(),
    'carat': 'std',
})

print(result)
```
