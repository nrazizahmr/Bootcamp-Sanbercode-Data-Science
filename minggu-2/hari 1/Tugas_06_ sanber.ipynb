<h2>Soal 3: DataFrame vs Series</h2>

- Apa perbedaan Series dengan list dan dictionary
- Apa perbedaan dataframe dan series
- Buatlah suatu data series

Jawab:



*   Sumber data: Series dengan list dibuat dengan data yang terurut dalam list, dan index ditentukan secara eksplisit oleh pengguna. Series dengan dictionary dibuat dengan data dalam bentuk pasangan key-value, di mana key menjadi index dan value menjadi data dalam Series. Index: Pada Series dengan list, index ditentukan secara manual oleh pengguna sesuai kebutuhan. Pada Series dengan dictionary, index secara otomatis diambil dari key dictionary, yang lebih praktis ketika data sudah berbentuk key value.
*   Series adalah struktur data satu dimensi. Mirip seperti satu kolom data dalam sebuah tabel, dengan setiap elemen memiliki index bawaan. Sedangkan DataFrame adalah struktur data dua dimensi yang bisa dianggap seperti tabel Excel. DataFrame memiliki baris dan kolom, yang memungkinkan Anda untuk menyimpan dan bekerja dengan data yang lebih kompleks.



```python
# code here
import pandas as pd

data_list = [10, 20, 30, 40]
series_list = pd.Series(data_list, index=['A', 'B', 'C', 'D'])
print("Series dari List:")
print(series_list)

data_dict = {'A': 100, 'B': 200, 'C': 300}
series_dict = pd.Series(data_dict)
print("\nSeries dari Dictionary:")
print(series_dict)
```

- baris pertama dari column price

<h2>Soal 5: Deskriptif Statistik</h2>

Jelaskan apa itu deskriptis statistik!

Jawab:

 Deskriptif Statistik adalah metode yang digunakan untuk menggambarkan data yang telah terkumpul, tanpa bermaksud membuat kesimpulan yang berlaku secara umum. Statistik deskriptif dapat digunakan untuk menyajikan data dalam bentuk tabel, diagram, grafik, atau besaran lainnya. Dapat dipecah menjadi ukuran kecenderungan sentral dan ukuran variabilitas. Ukuran kecenderungan sentral meliputi mean, median, dan modus. Sedangkan ukuran variabilitas meliputi deviasi standar, varians, variabel minimum dan maksimum, kurtosis, dan kemiringan.

<h2>Soal 6: Practice Deskriptif Statistik</h2>

- Gunakan suatu metode dari pandas untuk mengetahui deskriptif statistik dari suatu data
- Berapa nilai rata-rata dari column price
- Berapa nilai standar deviasi dari column depth
- Berapa nilai maximum dari column carat

```python
# code here
import pandas as pd

df = pd.read_csv('diamonds.csv')
desc_stats = df.describe()
print("Deskriptif Statistik:\n", desc_stats)

mean_price = df['price'].mean()
print("\nRata-rata dari kolom price:", mean_price)

std_depth = df['depth'].std()
print("\nStandar deviasi dari kolom depth:", std_depth)

max_carat = df['carat'].max()
print("\nNilai maksimum dari kolom carat:", max_carat)
```
