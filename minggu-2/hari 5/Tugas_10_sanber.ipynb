**Soal 4. Time grouping**



*   Lakukan grouping data perubahan iklim pada tahun 2016 berdasarkan frekuensi 1 bulan dengan memberikan agregasi max pada kolom meantemp. humidity, wind_speed dan meanpressure!
*   Berikan analisa, di bulan apa rata-rata temperatur, humidity, wind_speed dan meanpressure tertinggi?





```python
#code here
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)

df_2016 = df.loc['2016']

# grouping
monthly_max = df_2016.resample('ME').max()
print(monthly_max)

#agregasi max
max_meantemp_month = monthly_max['meantemp'].idxmax()
max_humidity_month = monthly_max['humidity'].idxmax()
max_wind_speed_month = monthly_max['wind_speed'].idxmax()
max_meanpressure_month = monthly_max['meanpressure'].idxmax()

print("\nAnalisis:")
print(f"- Temperatur tertinggi di bulan {max_meantemp_month.strftime('%B')} dengan nilai {monthly_max['meantemp'].max():.2f}.")
print(f"- Humidity tertinggi di bulan {max_humidity_month.strftime('%B')} dengan nilai {monthly_max['humidity'].max():.2f}.")
print(f"- Wind speed tertinggi di bulan {max_wind_speed_month.strftime('%B')} dengan nilai {monthly_max['wind_speed'].max():.2f}.")
print(f"- Mean pressure tertinggi di bulan {max_meanpressure_month.strftime('%B')} dengan nilai {monthly_max['meanpressure'].max():.2f}.")
```

**Soal 5. Time Differenced**

Tunjukkanlah perubahan iklim di tahun 2016 pada minggu terakhir

```python
#code here
df = pd.read_csv('climate_change.csv')
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)

last_week_2016 = df.loc['2016-12-26':'2016-12-31']
daily_diff = last_week_2016[['meantemp', 'humidity', 'wind_speed', 'meanpressure']].diff()
print("\nPerubahan  iklim di tahun 2016 pada minggu terakhir: \n")
print(daily_diff)
```
