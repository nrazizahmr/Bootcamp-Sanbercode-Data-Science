<h2>Soal 2: Data Restructuring</h2>

Jelaskan itu pivoting dan melting data



*   Pivoting adalah proses mengubah data dari format panjang (long format) menjadi format lebar (wide format). Ini biasanya dilakukan untuk merangkum data dan mengorganisasinya dalam bentuk tabel yang lebih mudah dipahami atau dianalisis.
*   Melting adalah kebalikan dari pivoting. Ini adalah proses mengubah data dari format lebar (wide format) menjadi format panjang (long format). Melting berguna ketika ingin mengonversi data menjadi bentuk yang lebih fleksibel dan siap untuk analisis atau visualisasi lebih lanjut.

<h2>Soal 3: Reshaping Data 1</h2>

```python
import pandas as pd
import random

data = {
    'Tipe Ujian' : ['UTS1', 'UAS1', 'UTS2', 'UAS2']*3,
    'Mata Pelajaran' : ['Fisika']*4 + ['Matematika']*4+['Statistika']*4,
    'Rata2 Waktu Belajar (menit)' : [random.choice(range(200, 800)) for i in range(12)],
    'Rata2 Nilai Ujian' : [random.choice(range(40, 100)) for i in range(12)]
}

df_1 = pd.DataFrame(data)
df_1
```

Perhatikan data di atas, kemudian ubahlah bentuknya agar sesuai yang di harapkan.

```python
#code here
pivot_df = pd.pivot_table(
    df_1,
    index='Tipe Ujian',
    columns='Mata Pelajaran',
    values=['Rata2 Nilai Ujian', 'Rata2 Waktu Belajar (menit)'],
    aggfunc='mean'
)

pivot_df.columns = [' '.join(col).strip() for col in pivot_df.columns]
pivot_df.reset_index(inplace=True)


print(pivot_df)
```

<h2>Soal 4: Reshaping Data 2</h2>

```python
import pandas as pd

data = {
    'Hari' : ['Sabtu', 'Minggu'],
    'Kebun Binatang' : [271, 399],
    'Dufan' : [501, 700],
    'Disney' : [1000, 1001],
    'Bali' : [900, 803]
}
df_2 = pd.DataFrame(data).set_index('Hari')
df_2
```

diatas adalah data tentang jumlah pengunjung di berbagai tempat hiburan, ubahlah struktur data tersebut sehingga sesuai dengan hasil yang diharapkan.

```python
#code here
reshaped_df = df_2.stack().reset_index()
reshaped_df.columns = ['Hari', 'Tempat Hiburan', 'Pengunjung']

print(reshaped_df)
```

<h2>Soal 5: Data Enrichment</h2>

Lakukan binning data menggunakan qcut sehingga sesuai hasil yang diharapkan

```python
#code here
df_1['Kategori Nilai'] = pd.qcut(df_1['Rata2 Nilai Ujian'], q=3, labels=['rendah', 'sedang', 'tinggi'])
def kelulusan(nilai):
    if nilai >= 75:
        return 'lulus'
    elif nilai >= 60:
        return 'remedial'
    else:
        return 'tidak lulus'

df_1['Keterangan Kelulusan'] = df_1['Rata2 Nilai Ujian'].apply(kelulusan)
print(df_1)
```
